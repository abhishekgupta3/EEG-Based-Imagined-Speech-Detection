### Data Acquisition: Person 1

```python
raw_data = mne.io.read_raw_cnt("Dataset/MM05/Acquisition 232 Data.cnt", preload=True).load_data()
```

```python
raw_data.get_data().shape
```

```python
raw_data.plot()
```

```python
raw_data.pick_channels(['FC6', 'FT8', 'C5', 'CP3', 'P3', 'T7', 'CP5', 'C3', 'CP1', 'C4'])
```

```python
raw_data.plot()
```

```python
raw_data = raw_data.filter(4, 40, n_jobs=4)
```

```python
raw_data.plot()
```

```python
raw = raw_data.get_data()
raw.shape
```

```python
numpy.save("Data/person1.npy", raw)
```
